# Exploratory Demo

This notebook is a lightweight companion to the repository. It reads the generated synthetic signals and scored outputs, then plots the risk score and inspects a few alert rows.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
data = pd.read_csv(root / 'data' / 'synthetic_signals.csv', parse_dates=['date'])
preds = pd.read_csv(root / 'outputs' / 'sample_predictions.csv', parse_dates=['date'])
data.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(preds['date'], preds['risk_score'], color='#0f766e', label='risk score')
ax.scatter(preds.loc[preds['alert_flag'] == 1, 'date'], preds.loc[preds['alert_flag'] == 1, 'risk_score'], color='#b91c1c', s=18, label='alert')
ax.set_title('Synthetic public-health signal risk score')
ax.set_xlabel('date')
ax.set_ylabel('risk score')
ax.grid(alpha=0.25)
ax.legend(frameon=False)
plt.show()
```
